## Стандартная ошибка регрессии:
$$
SER=\sqrt{\frac{RSS}{(n-k-1)}}
$$

## => RSS:
$$
RSS=(n-k-1)*SER^2
$$

```python
RSS=(n-k-1)*(SER**2) # RSS для 1-й регрессии
round(RSS,3)
```

## 1.2.8 Задача

```python
n=935 # observations
k=5 
SER=0.383
R2=0.178
```

## R^2:
$$
R^2=1-\frac{RSS}{TSS}
$$

## => TSS:
$$
TSS = \frac{RSS}{1-R^2}=\frac{(n-k-1)*SER^2}{1-R^2}
$$

```python
TSS=((n-k-1)*SER**2)/(1-R2) # TSS для 3-й регрессии
round(TSS,3)
```

## 1.2.9 Задача

```python
n=935 # observations
k=3 
SER=0.389
R2=0.149
```

## Стандартная ошибка регрессии:
$$
SER=\sqrt{\frac{RSS}{(n-k-1)}}
$$

## => RSS:
$$
RSS=(n-k-1)*SER^2
$$

```python
RSS=(n-k-1)*(SER**2) # RSS для 5-й регрессии
round(RSS,3)
```

## R^2:
$$
R^2=1-\frac{RSS}{TSS}
$$

## => TSS:
$$
TSS = \frac{RSS}{1-R^2}=\frac{(n-k-1)*SER^2}{1-R^2}
$$

```python
TSS=((n-k-1)*SER**2)/(1-R2) # TSS для 3-й регрессии
round(TSS,3)
```

## ESS:
$$
ESS = TSS - RSS
$$

```python
ESS=TSS-RSS
round(ESS,3)
```

## 2.1 Задача

```python
n=526 ## observations
k=5 ## exper, exper^2, female, smsa, south
R2=0.288
```

$$
R^2{adj}=1-(1-R^2)*\frac{(n-1)}{(n-k)}

```python
R2adj=1-(1-R2)*(n-1)/(n-k)
round(R2adj,3)
```
